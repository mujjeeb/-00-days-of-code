##Private Auctiion
Program takes down names of biddders and their bids, and then chooses the winner. This is done privately as bidders can't see other bidders bid.




```python
import os
logo = '''
                         ___________
                         \         /
                          )_______(
                          |"""""""|_.-._,.---------.,_.-._
                          |       | | |               | | ''-.
                          |       |_| |_             _| |_..-'
                          |_______| '-' `'---------'` '-'
                          )"""""""(
                         /_________\\
                       .-------------.
                      /_______________\\
'''
print(logo)
print("Welcome to the secret auction program")
bidders = {}
name = input("What is your name?: ")
bid = int(input("What is your bid?: $"))
other = input("Are there any other bidders? Type 'yes' or 'no'. " )
bidders [name] = bid


while other == 'yes' :
    os.system('cls' if os.name == 'nt' else 'clear')
    name = input("What is your name?: ")
    bid = int(input("What is your bid?: $"))
    other = input("Are there any other bidders? Type 'yes' or 'no'. " )
    bidders [name] = bid

#print(bidders)

for key in bidders:
    winner = 0
    winner_key = ""
    if bidders[key] > winner:
        winner = bidders[key]
        winner_key = key
   
       

print(f"The winner is {winner_key} with a bid of ${winner}")








```
